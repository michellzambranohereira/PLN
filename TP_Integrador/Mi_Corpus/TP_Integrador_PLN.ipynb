# Sección 2: Preprocesamiento

```python
#2.1 Carga y organización de datos
import pandas as pd
import numpy as np
import pickle
import re
import string
from collections import Counter
import nltk
import spacy
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import seaborn as sns
!python -m spacy download es_core_news_md

```

2.2 Limpieza de texto:

```python
# - Aplicar técnicas vistas en clase:
#   * Conversión a minúsculas
#   * Eliminación de signos de puntuación
#   * Eliminación de números (si no son relevantes)
#   * Eliminación de caracteres especiales
# - Justificar cada decisión de limpieza:

def limpiar_texto(texto):
    # Convertir a minúsculas
    texto = texto.lower() #Debido a que todas las letras están escritas en forma de versos, tendían a tener varias mayúsculas por linea, las cuales fueron elimindas

    #Eliminar signos de puntuación
    texto = re.sub(r"[^\w\s]", " ", texto) #La mayoría de las letras contenían signos como: ?¡!.,

    # Eliminación de números: no se realizó esta parte de limpieza ya que ninguna de las canciones tenía numeros en la letra.

    #Eliminar caracteres especiales y espacios extra
    texto = re.sub(r"\s+", " ", texto).strip() #se utilizó solo para eliminar las tabulaciones y espacios extras, ya que las páginas como Genius.com o Letras.com suele escribir las canciones en versos cortos, uno debajo del otro, para facilitar la lectura y el canto.

    return texto
```

```python
# Tomamos la primera letra del DataFrame
ejemplo_original = df.loc[0, "Letra"]
ejemplo_limpio = limpiar_texto(ejemplo_original)

print("🔹 Texto original:")
print(ejemplo_original[:300])   # primeros 300 caracteres

print("\n🔹 Texto limpio:")
print(ejemplo_limpio[:300])     # primeros 300 caracteres ya procesados
```

2.3 Tokenización y normalización:

```python
# - Separar texto en palabras
# - Decidir si aplicar stemming o lemmatización (justificar)
# - Mostrar ejemplos de antes y después

# Cargar el modelo de lenguaje en español
# Elegimos la lematización en lugar del stemming porque reduce las palabras a su forma base o de diccionario (lema),
# lo que suele generar tokens más significativos y conserva mejor el contexto de la palabra que el stemming.
# El stemming simplemente elimina sufijos y prefijos, lo que a veces puede crear palabras que no existen.


nlp = spacy.load("es_core_news_md") #Cargar el modelo de lenguaje en español

#A continuación función que tokeniza y lematiza el texto (letra) en español usando la librería spaCy.
#Devuelve una lista de lemas y se creó un dataframe para ver la comparación.
def tokenizar_y_lematizar(texto):

    doc = nlp(texto) #nlp(texto) crea un objeto doc, que contiene: tokens, lemas, etiquetas gramaticales, dependencias, etc.
    tokens_lematizados = [token.lemma_ for token in doc if not token.is_punct and not token.is_space]
    #Crea una lista de lemas de cada token
    # token.lemma_ = para la forma base de la palabra
    #if not token.is_punct and not token.is_space = descarta signos de puntuación y espacios,
    #a pesar de que en la sección 2.1 se realizó la limpieza del texto, este código filtra los tokens
    #descartando aquellos que presenten signos de puntuación o sean espacios en blanco o saltos de linea.

    comparativa = [(token.text, token.lemma_) for token in doc if not token.is_punct and not token.is_space]
    #Crea una lista de tuplas (original, lema) para cada token

    df_comparativa = pd.DataFrame(comparativa, columns=["Original", "Lema"])
    #Convierte esa lista en un dataframe con dos columnas que muestran la palabra original y la palabra lematizada

    return tokens_lematizados, df_comparativa
    #Devuelve la lista de lemas y el dataframe comparativo


#Creación de dos diccionarios vacíos:
textos_limpios_lematizados = {}
#textos_limpios_lematizados: guardará para cada texto (clave) la lista de lemas resultante.
comparativas_df = {}
#comparativas_df: guardará para cada texto (clave) un DataFrame con la comparación: palabra original vs lema.


# Iteramos sobre cada fila del DataFrame 'df', donde 'Título' es el nombre de la canción y 'Letra' es el texto de la canción.
for idx, row in df.iterrows():

    nombre = row["Título"]   # Se usa el título como clave (antes era el nombre del archivo)
    contenido = row["Letra"] # Ahora obtenemos directamente la letra desde el CSV

    # Aplicamos nuevamente la función de limpieza al contenido original.
    # Aunque ya se limpió antes para mostrar ejemplos, aquí se vuelve a aplicar porque:
    # - Se está procesando directamente desde 'Letra', que contiene los textos crudos.
    # - Asegura que todos los textos pasen por el mismo pipeline antes de lematizar.
    # - Esto garantiza consistencia y evita que se cuelen signos o espacios residuales.
    texto_limpio = limpiar_texto(contenido)

    # Aplicamos la función de tokenización y lematización al texto limpio.
    # Esta función devuelve: una lista de lemas (palabras en su forma base) y un dataframe comparativo.
    lemas, comparativa_df = tokenizar_y_lematizar(texto_limpio)

    #Guardamos la lista de lemas en el diccionario 'textos_limpios_lematizados', usando el título como clave.
    textos_limpios_lematizados[nombre] = lemas

    # Guardamos el DataFrame comparativo en el diccionario 'comparativas_df', también usando el título como clave.
    comparativas_df[nombre] = comparativa_df


#Código para mostrar el proceso de transformación de la letra desde su base original hasta la comparativa
# Filter out rows with NaN in 'Título' before sampling
df_filtered = df.dropna(subset=['Título'])
ejemplo_nombre = df_filtered.sample(1)["Título"].values[0] #Elección de una canción aleatoria
ejemplo_original = df_filtered.loc[df_filtered["Título"] == ejemplo_nombre, "Letra"].values[0] #Accede a la letra original
ejemplo_limpio = limpiar_texto(ejemplo_original) #Accede al texto ya limpio
ejemplo_lematizado = textos_limpios_lematizados[ejemplo_nombre] #Accede al texto lematizado
ejemplo_comparativa = comparativas_df[ejemplo_nombre] #Accede al dataframe donde está la comparativa entre el texto limpio y el texto lematizado

print(f"🔹 Texto original ('{ejemplo_nombre}'):")
print(ejemplo_original[:100])
print("\n🔹 Texto limpio:")
print(ejemplo_limpio[:100])
print("\n🔹 Tokens lematizados (primeros 50):")
print(ejemplo_lematizado[:50])
print("\n🔹 Comparativa palabra original vs lema (primeros 20):")
display(ejemplo_comparativa.head(20))
```

2.4 Manejo de stop words:

```python
nltk.download('stopwords')
nltk.download('punkt')

stopwords_es = set(stopwords.words('spanish')) #Carga del listado estándar de stop word en español, se usa set() para convertirlo en un conjutno y facilitar así la búsqueda.

#Elimina las stop words estándar en español de una lista de lemas.
def eliminar_stopwords(lemas):
    return [palabra for palabra in lemas if palabra not in stopwords_es]

# Aplicamos al ejemplo
ejemplo_filtrado = eliminar_stopwords(ejemplo_lematizado)

print("🔹 Tokens lematizados sin stop words (primeros 50):")
print(ejemplo_filtrado[:50])
```

- Justificar decisiones: Se utilizó el listado estándar de stop words en español por su confiabilidad y amplio uso en tareas de PLN. Esta elección simplifica el preprocesamiento y asegura consistencia en el análisis, sin personalizaciones adicionales en esta etapa.


2.5 Estadísticas post-procesamiento:

```python
# - Vocabulario final (número de palabras únicas)
# - Distribución de frecuencias

# Contar frecuencia de lemas sin stop words en todo el corpus
corpus_filtrado = []

# 🔹 Diccionario de correcciones para lematización defectuosa
correcciones = {
    "luno": "luna",
}

for nombre in textos_limpios_lematizados:
    lemas = textos_limpios_lematizados[nombre]
    lemas_filtrados = eliminar_stopwords(lemas)

    # 🔹 Arreglar tokens pegados y corregir lemas defectuosos
    tokens_arreglados = []
    for token in lemas_filtrados:
        # aplicar corrección si existe
        token = correcciones.get(token, token)
        # dividir si viene con espacios raros
        tokens_arreglados.extend(token.split())

    corpus_filtrado.extend(tokens_arreglados)

# Frecuencia total
frecuencia_total = Counter(corpus_filtrado)

# Vocabulario final
vocabulario_final = set(corpus_filtrado)

print(f"Número total de tokens (sin stop words): {len(corpus_filtrado)}")
print(f"Vocabulario final (palabras únicas): {len(vocabulario_final)}")
print("Top 20 palabras más frecuentes:")
print(frecuencia_total.most_common(20))

```

```python
# - Comparación antes/después del preprocesamiento

#Contadores totales
tokens_antes = 0
tokens_despues = 0

#Recorrer cada texto del corpus
for nombre, lemas in textos_limpios_lematizados.items():
    tokens_antes += len(lemas)  # Total de lemas antes de eliminar stop words
    tokens_despues += len(eliminar_stopwords(lemas))  # Total después

#Preparar datos para el gráfico
etapas = ['Antes del preprocesamiento', 'Después del preprocesamiento']
valores = [tokens_antes, tokens_despues]

# Graficar
plt.figure(figsize=(8, 5))
plt.bar(etapas, valores, color=['skyblue', 'salmon'])
plt.title('Comparación de cantidad de tokens antes y después del preprocesamiento')
plt.ylabel('Cantidad de tokens')
plt.tight_layout()
plt.show()
```

```python
#Unir todos los lemas del corpus (sin stop words)
corpus_filtrado = []

for nombre, lemas in textos_limpios_lematizados.items():
    lemas_sin_stopwords = eliminar_stopwords(lemas)
    corpus_filtrado.extend(lemas_sin_stopwords)

#Contar frecuencia de cada palabra
frecuencia = Counter(corpus_filtrado)
top_palabras = frecuencia.most_common(20)  # Top 20 más frecuentes

#Preparar datos para graficar
palabras = [palabra for palabra, _ in top_palabras]
frecuencias = [conteo for _, conteo in top_palabras]

#Graficar
plt.figure(figsize=(12, 6))
sns.barplot(x=frecuencias, y=palabras, hue=palabras, dodge=False, legend=False, palette="magma")
plt.title("Top 20 palabras más frecuentes en el corpus (sin stop words)", fontsize=14)
plt.xlabel("Frecuencia de las palabras")
plt.ylabel("Palabra")
plt.tight_layout()
plt.show()
```

# Sección 3: Análisis con BoW/TF-IDF

3.1 Vectorización del corpus:

```python
#Librerías a utilizar en esta sección
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# - Crear matriz documento-término con CountVectorizer
# - Crear matriz TF-IDF
# - Mostrar dimensiones y características de las matrices
# - Explicar qué representan los números

count_vectorizer = CountVectorizer(
    vocabulary=list(vocabulario_final)  # fijamos las columnas = tu set de palabras
)
matriz_count = count_vectorizer.fit_transform(
    [" ".join(eliminar_stopwords(textos_limpios_lematizados[nombre]))
     for nombre in textos_limpios_lematizados]
)

# Crear TF-IDF usando el mismo vocabulario
tfidf_vectorizer = TfidfVectorizer(
    vocabulary=list(vocabulario_final)
)
matriz_tfidf = tfidf_vectorizer.fit_transform(
    [" ".join(eliminar_stopwords(textos_limpios_lematizados[nombre]))
     for nombre in textos_limpios_lematizados]
)

# Mostrar resultados
print("📌 Matriz CountVectorizer:")
print("Dimensiones:", matriz_count.shape)  # debería ser (20, 680)
print("Ejemplo de términos:", count_vectorizer.get_feature_names_out()[:20])

print("\n📌 Matriz TF-IDF:")
print("Dimensiones:", matriz_tfidf.shape)  # también (20, 680)
print("Ejemplo de términos:", tfidf_vectorizer.get_feature_names_out()[:20])
```

Explicación de lo que representan los números:

En la matriz CountVectorizer:
- Cada fila representa un documento (canción).
- Cada columna representa una palabra única del corpus.
- Los valores son el número de veces que aparece la palabra en esa canción.

En la matriz TF-IDF:
- Igual que arriba (documentos en filas y palabras en columnas).
- Pero los valores están normalizados con la métrica TF-IDF.
Esto significa que:

TF (Term Frequency): frecuencia de una palabra en el documento.

DF (Inverse Document Frequency): penaliza palabras muy frecuentes en todo el corpus.

Resultado: valores entre 0 y 1 que indican la "importancia relativa" de la palabra dentro de una canción, considerando todo el corpus.


3.2 Términos más frecuentes y distintivos:

```python
# - Top 20 palabras más frecuentes (BoW)
# - Top 20 términos con mayor TF-IDF
# - Comparar ambas listas: ¿qué diferencias ves?
# - Interpretación: ¿estos términos caracterizan bien tu corpus?


#Top 20 palabras más frecuentes (BoW)
top20_bow = frecuencia_total.most_common(20)
print("🔹 Top 20 palabras más frecuentes (BoW):")
for palabra, freq in top20_bow:
    print(f"{palabra}: {freq}")

#Top 20 términos con mayor TF-IDF
#Sacamos la media de los pesos TF-IDF de cada palabra en el corpus
import numpy as np

tfidf_array = matriz_tfidf.toarray()
tfidf_mean = np.mean(tfidf_array, axis=0)

#Creamos diccionario término → peso medio
terminos = tfidf_vectorizer.get_feature_names_out()
tfidf_scores = dict(zip(terminos, tfidf_mean))

#Ordenamos y seleccionamos top 20
top20_tfidf = sorted(tfidf_scores.items(), key=lambda x: x[1], reverse=True)[:20]

print("\n🔹 Top 20 términos con mayor TF-IDF:")
for palabra, score in top20_tfidf:
    print(f"{palabra}: {score:.4f}")

#Comparación
print("\n📊 Comparación:")
print("Palabras que aparecen en ambos tops:", set([p for p,_ in top20_bow]) & set([p for p,_ in top20_tfidf]))
print("Palabras exclusivas de BoW:", set([p for p,_ in top20_bow]) - set([p for p,_ in top20_tfidf]))
print("Palabras exclusivas de TF-IDF:", set([p for p,_ in top20_tfidf]) - set([p for p,_ in top20_bow]))
```

- Comparar ambas listas: ¿qué diferencias ves?

BoW destaca palabras comunes como dar, ver, mujer, corazón, que reflejan el vocabulario compartido del merengue amoroso. TF-IDF, en cambio, resalta términos más específicos como vicio, ajeno, enamorar, que definen el estilo particular de ciertas canciones.

- Interpretación: ¿estos términos caracterizan bien tu corpus?

Ambos métodos caracterizan bien el corpus:

BoW muestra el lenguaje emocional general del género.

TF-IDF revela matices líricos y temáticos únicos por canción.

Juntos permiten entender tanto la base común como la diversidad estilística del merengue.

3.3 Matriz de similitud entre documentos

```python
# - Calcular similitud coseno entre todos los pares de documentos
# - Encontrar los 2-3 pares más similares
# - Encontrar los 2-3 pares más diferentes
# - Analizar: ¿tiene sentido lo que encontraste?

# ✅ Calcular similitud coseno entre todos los pares de documentos
similitud = cosine_similarity(matriz_tfidf)

# Guardamos nombres de los documentos (canciones)
nombres_docs = list(textos_limpios_lematizados.keys())

# Convertimos en matriz (DataFrame) para verlo mejor
import pandas as pd
df_similitud = pd.DataFrame(similitud, index=nombres_docs, columns=nombres_docs)

print("🔹 Matriz de similitud coseno:")
print(df_similitud.round(3))

# ✅ Encontrar los 2-3 pares más similares (excluyendo diagonal = 1.0)
similitudes_superiores = []
for i in range(len(nombres_docs)):
    for j in range(i+1, len(nombres_docs)):
        similitudes_superiores.append(((nombres_docs[i], nombres_docs[j]), similitud[i, j]))

# Ordenamos de mayor a menor similitud
similitudes_ordenadas = sorted(similitudes_superiores, key=lambda x: x[1], reverse=True)

```

```python
print("\n🔹 PARES MÁS SIMILARES:")
for par, score in similitudes_ordenadas[:3]:
    print(f"{par}: {score:.3f}")

# ✅ Encontrar los 2-3 pares más diferentes
print("\n🔹 PARES MÁS DIFERENTES:")
for par, score in similitudes_ordenadas[-3:]:
    print(f"{par}: {score:.3f}")
```

- Analizar: ¿tiene sentido lo que encontraste?

No precisamente, las cancionces que están en 'pares más similares', en cuanto a tono emocional y sentido, son diferentes entre ellas.

# Sección 4: Análisis con Word Embeddings

4.1 Aplicación de embeddings:

```python
# Cargar modelo en español
nlp = spacy.load("es_core_news_md")

# 4.1 Aplicación de embeddings
# - Procesar tu corpus con spaCy
# - Obtener vectores para documentos (promedio de vectores de palabras)

# Función para obtener el vector promedio de una canción
def vector_promedio(lemas):
    doc = nlp(" ".join(lemas))
    return doc.vector

# Crear diccionario de vectores
vectores_documentos = {
    nombre: vector_promedio(lemas)
    for nombre, lemas in textos_limpios_lematizados.items()
}

# Lista de vectores en el mismo orden que los títulos
nombres_docs = list(vectores_documentos.keys())
matriz_embeddings = np.array([vectores_documentos[nombre] for nombre in nombres_docs])

print("📐 Dimensiones de la matriz de embeddings:", matriz_embeddings.shape)
```

- Explicar qué son los embeddings y por qué son diferentes a BoW

A diferencia de los enfoques anteriores (BoW y TF-IDF), los embeddings permiten representar cada documento como un vector que captura el significado contextual de sus palabras. Para cada canción, se calcula el vector promedio de sus lemas, generando una matriz de dimensiones (20, 300), donde cada fila representa una canción y cada columna una dimensión semántica.

Esta representación será utilizada en las siguientes secciones para comparar similitud semántica entre canciones, buscar analogías y visualizar agrupamientos naturales.

4.2 Análisis de similitud semántica

```python
# - Calcular similitud entre documentos usando embeddings
# - Comparar con resultados de TF-IDF

from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# Calcular similitud coseno entre vectores de embeddings
similitud_embeddings = cosine_similarity(matriz_embeddings)

# Convertir a DataFrame para visualización
df_similitud_embeddings = pd.DataFrame(similitud_embeddings, index=nombres_docs, columns=nombres_docs)

# Mostrar matriz redondeada
df_similitud_embeddings.round(3)

# Extraer pares únicos (sin repetir ni incluir diagonal)
pares_similitud = []
for i in range(len(nombres_docs)):
    for j in range(i + 1, len(nombres_docs)):
        pares_similitud.append(((nombres_docs[i], nombres_docs[j]), similitud_embeddings[i, j]))

# Ordenar por similitud
pares_ordenados = sorted(pares_similitud, key=lambda x: x[1], reverse=True)

# Mostrar top 3 más similares
print("🔹 PARES MÁS SIMILARES (embeddings):")
for par, score in pares_ordenados[:3]:
    print(f"{par}: {score:.3f}")

# Mostrar top 3 más diferentes
print("\n🔹 PARES MÁS DIFERENTES (embeddings):")
for par, score in pares_ordenados[-3:]:
    print(f"{par}: {score:.3f}")
```

En comparación con los resultados de TF-IDF, vemos que tenemos diferencias con el resultado de embedding:

- TF-IDF mide coincidencias de palabras, como se dijo anteriormente, no capta el significado profundo ni tono.
- Embedding si capta el significado contextual de las palabras, pueden reflejar la carga semántica y emocional del texto. Por ejemplo la similitud entre 'Nuestra Canción'vs 'El Loco y La Luna': 0.957, es acertada, ambas canciones tienen como tema central el recuerdo de un amor que ya no está, con un tono emocional nostáligico y triste.
