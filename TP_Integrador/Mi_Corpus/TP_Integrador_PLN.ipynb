# Secci√≥n 2: Preprocesamiento

```python
#2.1 Carga y organizaci√≥n de datos
import pandas as pd
import numpy as np
import pickle
import re
import string
from collections import Counter
import nltk
import spacy
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import seaborn as sns
!python -m spacy download es_core_news_md

```

2.2 Limpieza de texto:

```python
# - Aplicar t√©cnicas vistas en clase:
#   * Conversi√≥n a min√∫sculas
#   * Eliminaci√≥n de signos de puntuaci√≥n
#   * Eliminaci√≥n de n√∫meros (si no son relevantes)
#   * Eliminaci√≥n de caracteres especiales
# - Justificar cada decisi√≥n de limpieza:

def limpiar_texto(texto):
    # Convertir a min√∫sculas
    texto = texto.lower() #Debido a que todas las letras est√°n escritas en forma de versos, tend√≠an a tener varias may√∫sculas por linea, las cuales fueron elimindas

    #Eliminar signos de puntuaci√≥n
    texto = re.sub(r"[^\w\s]", " ", texto) #La mayor√≠a de las letras conten√≠an signos como: ?¬°!.,

    # Eliminaci√≥n de n√∫meros: no se realiz√≥ esta parte de limpieza ya que ninguna de las canciones ten√≠a numeros en la letra.

    #Eliminar caracteres especiales y espacios extra
    texto = re.sub(r"\s+", " ", texto).strip() #se utiliz√≥ solo para eliminar las tabulaciones y espacios extras, ya que las p√°ginas como Genius.com o Letras.com suele escribir las canciones en versos cortos, uno debajo del otro, para facilitar la lectura y el canto.

    return texto
```

2.3 Tokenizaci√≥n y normalizaci√≥n:

```python
# - Separar texto en palabras
# - Decidir si aplicar stemming o lemmatizaci√≥n (justificar)
# - Mostrar ejemplos de antes y despu√©s

# Cargar el modelo de lenguaje en espa√±ol
# Elegimos la lematizaci√≥n en lugar del stemming porque reduce las palabras a su forma base o de diccionario (lema),
# lo que suele generar tokens m√°s significativos y conserva mejor el contexto de la palabra que el stemming.
# El stemming simplemente elimina sufijos y prefijos, lo que a veces puede crear palabras que no existen.


nlp = spacy.load("es_core_news_md") #Cargar el modelo de lenguaje en espa√±ol

#A continuaci√≥n funci√≥n que tokeniza y lematiza el texto (letra) en espa√±ol usando la librer√≠a spaCy.
#Devuelve una lista de lemas y se cre√≥ un dataframe para ver la comparaci√≥n.
def tokenizar_y_lematizar(texto):

    doc = nlp(texto) #nlp(texto) crea un objeto doc, que contiene: tokens, lemas, etiquetas gramaticales, dependencias, etc.
    tokens_lematizados = [token.lemma_ for token in doc if not token.is_punct and not token.is_space]
    #Crea una lista de lemas de cada token
    # token.lemma_ = para la forma base de la palabra
    #if not token.is_punct and not token.is_space = descarta signos de puntuaci√≥n y espacios,
    #a pesar de que en la secci√≥n 2.1 se realiz√≥ la limpieza del texto, este c√≥digo filtra los tokens
    #descartando aquellos que presenten signos de puntuaci√≥n o sean espacios en blanco o saltos de linea.

    comparativa = [(token.text, token.lemma_) for token in doc if not token.is_punct and not token.is_space]
    #Crea una lista de tuplas (original, lema) para cada token

    df_comparativa = pd.DataFrame(comparativa, columns=["Original", "Lema"])
    #Convierte esa lista en un dataframe con dos columnas que muestran la palabra original y la palabra lematizada

    return tokens_lematizados, df_comparativa
    #Devuelve la lista de lemas y el dataframe comparativo


#Creaci√≥n de dos diccionarios vac√≠os:
textos_limpios_lematizados = {}
#textos_limpios_lematizados: guardar√° para cada texto (clave) la lista de lemas resultante.
comparativas_df = {}
#comparativas_df: guardar√° para cada texto (clave) un DataFrame con la comparaci√≥n: palabra original vs lema.


# Iteramos sobre cada fila del DataFrame 'df', donde 'T√≠tulo' es el nombre de la canci√≥n y 'Letra' es el texto de la canci√≥n.
for idx, row in df.iterrows():

    nombre = row["T√≠tulo"]   # Se usa el t√≠tulo como clave (antes era el nombre del archivo)
    contenido = row["Letra"] # Ahora obtenemos directamente la letra desde el CSV

    # Aplicamos nuevamente la funci√≥n de limpieza al contenido original.
    # Aunque ya se limpi√≥ antes para mostrar ejemplos, aqu√≠ se vuelve a aplicar porque:
    # - Se est√° procesando directamente desde 'Letra', que contiene los textos crudos.
    # - Asegura que todos los textos pasen por el mismo pipeline antes de lematizar.
    # - Esto garantiza consistencia y evita que se cuelen signos o espacios residuales.
    texto_limpio = limpiar_texto(contenido)

    # Aplicamos la funci√≥n de tokenizaci√≥n y lematizaci√≥n al texto limpio.
    # Esta funci√≥n devuelve: una lista de lemas (palabras en su forma base) y un dataframe comparativo.
    lemas, comparativa_df = tokenizar_y_lematizar(texto_limpio)

    #Guardamos la lista de lemas en el diccionario 'textos_limpios_lematizados', usando el t√≠tulo como clave.
    textos_limpios_lematizados[nombre] = lemas

    # Guardamos el DataFrame comparativo en el diccionario 'comparativas_df', tambi√©n usando el t√≠tulo como clave.
    comparativas_df[nombre] = comparativa_df


#C√≥digo para mostrar el proceso de transformaci√≥n de la letra desde su base original hasta la comparativa
# Filter out rows with NaN in 'T√≠tulo' before sampling
df_filtered = df.dropna(subset=['T√≠tulo'])
ejemplo_nombre = df_filtered.sample(1)["T√≠tulo"].values[0] #Elecci√≥n de una canci√≥n aleatoria
ejemplo_original = df_filtered.loc[df_filtered["T√≠tulo"] == ejemplo_nombre, "Letra"].values[0] #Accede a la letra original
ejemplo_limpio = limpiar_texto(ejemplo_original) #Accede al texto ya limpio
ejemplo_lematizado = textos_limpios_lematizados[ejemplo_nombre] #Accede al texto lematizado
ejemplo_comparativa = comparativas_df[ejemplo_nombre] #Accede al dataframe donde est√° la comparativa entre el texto limpio y el texto lematizado

print(f"üîπ Texto original ('{ejemplo_nombre}'):")
print(ejemplo_original[:100])
print("\nüîπ Texto limpio:")
print(ejemplo_limpio[:100])
print("\nüîπ Tokens lematizados (primeros 50):")
print(ejemplo_lematizado[:50])
print("\nüîπ Comparativa palabra original vs lema (primeros 20):")
display(ejemplo_comparativa.head(20))
```

2.4 Manejo de stop words:

```python
nltk.download('stopwords')
nltk.download('punkt')

stopwords_es = set(stopwords.words('spanish')) #Carga del listado est√°ndar de stop word en espa√±ol, se usa set() para convertirlo en un conjutno y facilitar as√≠ la b√∫squeda.

#Elimina las stop words est√°ndar en espa√±ol de una lista de lemas.
def eliminar_stopwords(lemas):
    return [palabra for palabra in lemas if palabra not in stopwords_es]

# Aplicamos al ejemplo
ejemplo_filtrado = eliminar_stopwords(ejemplo_lematizado)

print("üîπ Tokens lematizados sin stop words (primeros 50):")
print(ejemplo_filtrado[:50])
```

```python
print(f"üî¢ Total antes: {len(ejemplo_lematizado)}")
print(f"üî¢ Total despu√©s de eliminar stop words: {len(ejemplo_filtrado)}")
```

- Decidir qu√© stop words usar (espa√±ol est√°ndar + espec√≠ficas del dominio): Se utiliza el listado est√°ndar de stop words en espa√±ol (NLTK).

- Justificar decisiones: Se utiliz√≥ el listado est√°ndar de stop words en espa√±ol por su confiabilidad y amplio uso en tareas de PLN. Esta elecci√≥n simplifica el preprocesamiento y asegura consistencia en el an√°lisis, sin personalizaciones adicionales en esta etapa.


2.5 Estad√≠sticas post-procesamiento:

```python
# - Vocabulario final (n√∫mero de palabras √∫nicas)
# - Distribuci√≥n de frecuencias

# Contar frecuencia de lemas sin stop words en todo el corpus
corpus_filtrado = []

# üîπ Diccionario de correcciones para lematizaci√≥n defectuosa
correcciones = {
    "luno": "luna",
}

for nombre in textos_limpios_lematizados:
    lemas = textos_limpios_lematizados[nombre]
    lemas_filtrados = eliminar_stopwords(lemas)

    # üîπ Arreglar tokens pegados y corregir lemas defectuosos
    tokens_arreglados = []
    for token in lemas_filtrados:
        # aplicar correcci√≥n si existe
        token = correcciones.get(token, token)
        # dividir si viene con espacios raros
        tokens_arreglados.extend(token.split())

    corpus_filtrado.extend(tokens_arreglados)

# Frecuencia total
frecuencia_total = Counter(corpus_filtrado)

# Vocabulario final
vocabulario_final = set(corpus_filtrado)

print(f"N√∫mero total de tokens (sin stop words): {len(corpus_filtrado)}")
print(f"Vocabulario final (palabras √∫nicas): {len(vocabulario_final)}")
print("Top 20 palabras m√°s frecuentes:")
print(frecuencia_total.most_common(20))

```

```python
# - Comparaci√≥n antes/despu√©s del preprocesamiento

#Contadores totales
tokens_antes = 0
tokens_despues = 0

#Recorrer cada texto del corpus
for nombre, lemas in textos_limpios_lematizados.items():
    tokens_antes += len(lemas)  # Total de lemas antes de eliminar stop words
    tokens_despues += len(eliminar_stopwords(lemas))  # Total despu√©s

#Preparar datos para el gr√°fico
etapas = ['Antes del preprocesamiento', 'Despu√©s del preprocesamiento']
valores = [tokens_antes, tokens_despues]

# Graficar
plt.figure(figsize=(8, 5))
plt.bar(etapas, valores, color=['skyblue', 'salmon'])
plt.title('Comparaci√≥n de cantidad de tokens antes y despu√©s del preprocesamiento')
plt.ylabel('Cantidad de tokens')
plt.tight_layout()
plt.show()
```

```python
#Unir todos los lemas del corpus (sin stop words)
corpus_filtrado = []

for nombre, lemas in textos_limpios_lematizados.items():
    lemas_sin_stopwords = eliminar_stopwords(lemas)
    corpus_filtrado.extend(lemas_sin_stopwords)

#Contar frecuencia de cada palabra
frecuencia = Counter(corpus_filtrado)
top_palabras = frecuencia.most_common(20)  # Top 20 m√°s frecuentes

#Preparar datos para graficar
palabras = [palabra for palabra, _ in top_palabras]
frecuencias = [conteo for _, conteo in top_palabras]

#Graficar
plt.figure(figsize=(12, 6))
sns.barplot(x=frecuencias, y=palabras, hue=palabras, dodge=False, legend=False, palette="magma")
plt.title("Top 20 palabras m√°s frecuentes en el corpus (sin stop words)", fontsize=14)
plt.xlabel("Frecuencia de las palabras")
plt.ylabel("Palabra")
plt.tight_layout()
plt.show()
```

# Secci√≥n 3: An√°lisis con BoW/TF-IDF

3.1 Vectorizaci√≥n del corpus:

```python
#Librer√≠as a utilizar en esta secci√≥n
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# - Crear matriz documento-t√©rmino con CountVectorizer
# - Crear matriz TF-IDF
# - Mostrar dimensiones y caracter√≠sticas de las matrices
# - Explicar qu√© representan los n√∫meros

count_vectorizer = CountVectorizer(
    vocabulary=list(vocabulario_final)  # fijamos las columnas = tu set de palabras
)
matriz_count = count_vectorizer.fit_transform(
    [" ".join(eliminar_stopwords(textos_limpios_lematizados[nombre]))
     for nombre in textos_limpios_lematizados]
)

# Crear TF-IDF usando el mismo vocabulario
tfidf_vectorizer = TfidfVectorizer(
    vocabulary=list(vocabulario_final)
)
matriz_tfidf = tfidf_vectorizer.fit_transform(
    [" ".join(eliminar_stopwords(textos_limpios_lematizados[nombre]))
     for nombre in textos_limpios_lematizados]
)

# Mostrar resultados
print("üìå Matriz CountVectorizer:")
print("Dimensiones:", matriz_count.shape)  # deber√≠a ser (20, 680)
print("Ejemplo de t√©rminos:", count_vectorizer.get_feature_names_out()[:20])

print("\nüìå Matriz TF-IDF:")
print("Dimensiones:", matriz_tfidf.shape)  # tambi√©n (20, 680)
print("Ejemplo de t√©rminos:", tfidf_vectorizer.get_feature_names_out()[:20])
```

Explicaci√≥n de lo que representan los n√∫meros:

En la matriz CountVectorizer:
- Cada fila representa un documento (canci√≥n).
- Cada columna representa una palabra √∫nica del corpus.
- Los valores son el n√∫mero de veces que aparece la palabra en esa canci√≥n.

En la matriz TF-IDF:
- Igual que arriba (documentos en filas y palabras en columnas).
- Pero los valores est√°n normalizados con la m√©trica TF-IDF.
Esto significa que:

TF (Term Frequency): frecuencia de una palabra en el documento.

DF (Inverse Document Frequency): penaliza palabras muy frecuentes en todo el corpus.

Resultado: valores entre 0 y 1 que indican la "importancia relativa" de la palabra dentro de una canci√≥n, considerando todo el corpus.


3.2 T√©rminos m√°s frecuentes y distintivos:

```python
# - Top 20 palabras m√°s frecuentes (BoW)
# - Top 20 t√©rminos con mayor TF-IDF
# - Comparar ambas listas: ¬øqu√© diferencias ves?
# - Interpretaci√≥n: ¬øestos t√©rminos caracterizan bien tu corpus?


#Top 20 palabras m√°s frecuentes (BoW)
top20_bow = frecuencia_total.most_common(20)
print("üîπ Top 20 palabras m√°s frecuentes (BoW):")
for palabra, freq in top20_bow:
    print(f"{palabra}: {freq}")

#Top 20 t√©rminos con mayor TF-IDF
#Sacamos la media de los pesos TF-IDF de cada palabra en el corpus
import numpy as np

tfidf_array = matriz_tfidf.toarray()
tfidf_mean = np.mean(tfidf_array, axis=0)

#Creamos diccionario t√©rmino ‚Üí peso medio
terminos = tfidf_vectorizer.get_feature_names_out()
tfidf_scores = dict(zip(terminos, tfidf_mean))

#Ordenamos y seleccionamos top 20
top20_tfidf = sorted(tfidf_scores.items(), key=lambda x: x[1], reverse=True)[:20]

print("\nüîπ Top 20 t√©rminos con mayor TF-IDF:")
for palabra, score in top20_tfidf:
    print(f"{palabra}: {score:.4f}")

#Comparaci√≥n
print("\nüìä Comparaci√≥n:")
print("Palabras que aparecen en ambos tops:", set([p for p,_ in top20_bow]) & set([p for p,_ in top20_tfidf]))
print("Palabras exclusivas de BoW:", set([p for p,_ in top20_bow]) - set([p for p,_ in top20_tfidf]))
print("Palabras exclusivas de TF-IDF:", set([p for p,_ in top20_tfidf]) - set([p for p,_ in top20_bow]))
```

- Comparar ambas listas: ¬øqu√© diferencias ves?

BoW destaca palabras comunes como dar, ver, mujer, coraz√≥n, que reflejan el vocabulario compartido del merengue amoroso. TF-IDF, en cambio, resalta t√©rminos m√°s espec√≠ficos como vicio, ajeno, enamorar, que definen el estilo particular de ciertas canciones.

- Interpretaci√≥n: ¬øestos t√©rminos caracterizan bien tu corpus?

Ambos m√©todos caracterizan bien el corpus:

BoW muestra el lenguaje emocional general del g√©nero.

TF-IDF revela matices l√≠ricos y tem√°ticos √∫nicos por canci√≥n.

# Secci√≥n 4: An√°lisis con Word Embeddings

4.1 Aplicaci√≥n de embeddings:

```python
# Cargar modelo en espa√±ol
nlp = spacy.load("es_core_news_md")

# 4.1 Aplicaci√≥n de embeddings
# - Procesar tu corpus con spaCy
# - Obtener vectores para documentos (promedio de vectores de palabras)

# Funci√≥n para obtener el vector promedio de una canci√≥n
def vector_promedio(lemas):
    doc = nlp(" ".join(lemas))
    return doc.vector

# Crear diccionario de vectores
vectores_documentos = {
    nombre: vector_promedio(lemas)
    for nombre, lemas in textos_limpios_lematizados.items()
}

# Lista de vectores en el mismo orden que los t√≠tulos
nombres_docs = list(vectores_documentos.keys())
matriz_embeddings = np.array([vectores_documentos[nombre] for nombre in nombres_docs])

print("üìê Dimensiones de la matriz de embeddings:", matriz_embeddings.shape)
```

- Explicar qu√© son los embeddings y por qu√© son diferentes a BoW

A diferencia de los enfoques anteriores (BoW y TF-IDF), los embeddings permiten representar cada documento como un vector que captura el significado contextual de sus palabras. Para cada canci√≥n, se calcula el vector promedio de sus lemas, generando una matriz de dimensiones (20, 300), donde cada fila representa una canci√≥n y cada columna una dimensi√≥n sem√°ntica.

Esta representaci√≥n ser√° utilizada en las siguientes secciones para comparar similitud sem√°ntica entre canciones, buscar analog√≠as y visualizar agrupamientos naturales.

4.2 An√°lisis de similitud sem√°ntica

```python
# - Calcular similitud entre documentos usando embeddings
# - Comparar con resultados de TF-IDF

from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# Calcular similitud coseno entre vectores de embeddings
similitud_embeddings = cosine_similarity(matriz_embeddings)

# Convertir a DataFrame para visualizaci√≥n
df_similitud_embeddings = pd.DataFrame(similitud_embeddings, index=nombres_docs, columns=nombres_docs)

# Mostrar matriz redondeada
df_similitud_embeddings.round(3)

# Extraer pares √∫nicos (sin repetir ni incluir diagonal)
pares_similitud = []
for i in range(len(nombres_docs)):
    for j in range(i + 1, len(nombres_docs)):
        pares_similitud.append(((nombres_docs[i], nombres_docs[j]), similitud_embeddings[i, j]))

# Ordenar por similitud
pares_ordenados = sorted(pares_similitud, key=lambda x: x[1], reverse=True)

# Mostrar top 3 m√°s similares
print("üîπ PARES M√ÅS SIMILARES (embeddings):")
for par, score in pares_ordenados[:3]:
    print(f"{par}: {score:.3f}")

# Mostrar top 3 m√°s diferentes
print("\nüîπ PARES M√ÅS DIFERENTES (embeddings):")
for par, score in pares_ordenados[-3:]:
    print(f"{par}: {score:.3f}")
```

En comparaci√≥n con los resultados de TF-IDF, vemos que tenemos diferencias con el resultado de embedding:

- TF-IDF mide coincidencias de palabras, como se dijo anteriormente, no capta el significado profundo ni tono.
- Embedding si capta el significado contextual de las palabras, pueden reflejar la carga sem√°ntica y emocional del texto. Por ejemplo la similitud entre 'Nuestra Canci√≥n'vs 'El Loco y La Luna': 0.957, es acertada, ambas canciones tienen como tema central el recuerdo de un amor que ya no est√°, con un tono emocional nost√°ligico y triste.


4.3 B√∫squeda de analog√≠as relevantes al corpus:

```python
# - Encontrar palabras m√°s similares a t√©rminos clave de tu dominio
# - Intentar crear 2-3 analog√≠as que funcionen con vocabulario de tu corpus
# - Ejemplo: "rock es a guitarra como tango es a ?"
# - Interpretar: ¬ølas analog√≠as tienen sentido?


# Buscar palabras similares a "amor"
amor = nlp.vocab["amor"]
similares_amor = sorted(
    amor.vocab,
    key=lambda w: w.similarity(amor) if w.has_vector else -1,
    reverse=True
)

# Mostrar top 10 similares
print("üîç Palabras m√°s similares a 'amor':")
for w in similares_amor[:10]:
    print(w.text, round(w.similarity(amor), 3))

# Ejemplo: 'vida' es a 'coraz√≥n' como 'alma' es a
vida = nlp.vocab["vida"].vector
coraz√≥n = nlp.vocab["coraz√≥n"].vector
alma = nlp.vocab["alma"].vector

# Vector resultante
resultado = vida - coraz√≥n + alma

# Buscar palabra m√°s cercana al resultado
mejor = None
mejor_similitud = -1
for palabra in nlp.vocab:
    if palabra.has_vector and palabra.is_alpha:
        similitud = palabra.vector @ resultado
        if similitud > mejor_similitud:
            mejor = palabra
            mejor_similitud = similitud

print("üß† Analog√≠a: 'vida' es a 'coraz√≥n' como 'alma' es a ‚Üí", mejor.text)
```
